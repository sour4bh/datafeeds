```python
def get_pid(url):
    try:
        parsed_url = urlparse(url)
        qs_prep = parse_qs(parsed_url.query)["destinationUrl"]
        product_id = qs_prep[0].split("/")[-2]
        return product_id
    except Exception:
        return None

df["pid"] = df.merchant_deep_link.apply(get_pid)
df
```
